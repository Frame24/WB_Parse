```python
import torch
from transformers import BertTokenizer, BertModel, BertForSequenceClassification
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from tqdm import tqdm
import numpy as np
import logging
import nltk

#    NLTK
nltk.download('stopwords')
nltk.download('wordnet')

#  
log_file_path = './reviews_keywords/logfile.log'
logging.basicConfig(filename=log_file_path, level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

#     
model_name = 'DeepPavlov/rubert-base-cased'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)
sentiment_model = BertForSequenceClassification.from_pretrained('./reviews_keywords/results/checkpoint-235')

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
sentiment_model.to(device)

#  -   
stop_words = set(stopwords.words('russian'))
lemmatizer = WordNetLemmatizer()

#     (   -)
def normalize_text(text):
    tokens = [lemmatizer.lemmatize(word) for word in text.split() if word not in stop_words]
    return ' '.join(tokens)

#     
def get_embedding(text):
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=128)
    inputs = {key: val.to(device) for key, val in inputs.items()}
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).cpu().numpy().flatten()

#    
def normalize_vector(vector):
    norm = np.linalg.norm(vector)
    return vector / norm if norm != 0 else vector

#  BERT  TF-IDF     
def combine_bert_tfidf(text, tfidf_vectorizer, tfidf_features):
    bert_embedding = get_embedding(text)
    tfidf_vector = tfidf_vectorizer.transform([text]).toarray().flatten()
    combined_vector = np.concatenate((bert_embedding, tfidf_vector), axis=0)
    return combined_vector

#        
def find_best_groups(category, aspect_groups, tfidf_vectorizer, tfidf_features, threshold=0.4):
    c_split = category.split("/")
    category = f"{c_split[1]} {c_split[-1]}".lower()
    category_vector = combine_bert_tfidf(category, tfidf_vectorizer, tfidf_features)
    best_similarity = 0
    best_group = None
    
    for group_name in aspect_groups.keys():
        group_vector = combine_bert_tfidf(group_name, tfidf_vectorizer, tfidf_features)
        similarity = np.dot(category_vector, group_vector)
        
        #  similarity score  
        logging.info(f"Category: {category} | Group: {group_name} | Similarity: {similarity}")
        
        if similarity > best_similarity and similarity >= threshold:
            best_similarity = similarity
            best_group = group_name
    
    return [' ', best_group] if best_group else [' ']

#     
def find_closest_aspect(embedding, aspect_embeddings, threshold=0.4):
    similarities = {}
    for aspect, aspect_emb in aspect_embeddings.items():
        similarity = cosine_similarity(embedding.reshape(1, -1), aspect_emb.reshape(1, -1)).flatten()[0]
        similarities[aspect] = similarity
    
    #    similarity   -3
    sorted_aspects = sorted(similarities.items(), key=lambda item: item[1], reverse=True)[:3]
    
    #  -3 
    for aspect, similarity in sorted_aspects:
        logging.info(f"Aspect: {aspect} | Similarity: {similarity}")
    
    #   ,   similarity  
    closest_aspect, similarity_score = sorted_aspects[0]
    if similarity_score >= threshold:
        return closest_aspect, similarity_score
    else:
        return "", similarity_score  #  ,    

#    
def predict_sentiment(text):
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=128).to(device)
    with torch.no_grad():
        outputs = sentiment_model(**inputs)
        sentiment = torch.argmax(outputs.logits, dim=-1).cpu().numpy()
    return 'positive' if sentiment == 1 else 'negative'

#   DataFrame
def process_reviews(df, aspect_groups_all, tfidf_vectorizer, tfidf_features):
    #    
    aspect_embeddings_cache = {}
    
    #    
    all_original_aspects = []
    all_matched_aspects = []
    all_similarity_scores = []
    all_sentiments = []

    categories = df['category'].unique()

    for category in tqdm(categories, desc=" "):
        best_groups = find_best_groups(category, aspect_groups_all, tfidf_vectorizer, tfidf_features)

        if category not in aspect_embeddings_cache:
            aspect_embeddings = {aspect: combine_bert_tfidf(aspect, tfidf_vectorizer, tfidf_features) for group in best_groups for aspect in aspect_groups_all[group].keys()}
            aspect_embeddings_cache[category] = aspect_embeddings
        
        category_df = df[df['category'] == category]
        
        for review in tqdm(category_df['processed_review'], desc=f"    {category}"):
            original_aspects = []
            matched_aspects = []
            similarity_scores = []
            sentiments = []

            phrases = review.split('. ')
            
            for phrase in phrases:
                normalized_phrase = normalize_text(phrase)
                phrase_embedding = combine_bert_tfidf(normalized_phrase, tfidf_vectorizer, tfidf_features)
                
                #   
                closest_aspect, similarity_score = find_closest_aspect(phrase_embedding, aspect_embeddings_cache[category])
                
                original_aspects.append(phrase)
                matched_aspects.append(closest_aspect)
                similarity_scores.append(similarity_score)
                sentiments.append(predict_sentiment(phrase))
            
            all_original_aspects.append(original_aspects)
            all_matched_aspects.append(matched_aspects)
            all_similarity_scores.append(similarity_scores)
            all_sentiments.append(sentiments)

    #    DataFrame
    df['original_aspects'] = all_original_aspects
    df['matched_aspects'] = all_matched_aspects
    df['similarity_score'] = all_similarity_scores
    df['predicted_sentiment'] = all_sentiments

    return df

#  TF-IDF Vectorizer   
def train_tfidf(df):
    processed_texts = df['processed_review'].apply(normalize_text)
    tfidf_vectorizer = TfidfVectorizer(max_features=5000)
    tfidf_features = tfidf_vectorizer.fit_transform(processed_texts)
    return tfidf_vectorizer, tfidf_features

#     DataFrame
start_idx = 3000
end_idx = 5000
test_df = df_raw.iloc[start_idx:end_idx].copy()
test_df['processed_review'] = test_df['review_full_text'].apply(preprocess_text)

#  TF-IDF  
tfidf_vectorizer, tfidf_features = train_tfidf(test_df)

#    
test_df = process_reviews(test_df, aspect_groups_all, tfidf_vectorizer, tfidf_features)

#  
output_path = 'path_to_save_results.csv'
test_df.to_csv(output_path, index=False)
print(f"   {output_path}")

#      
display(test_df[['review_full_text', 'processed_review', 'original_aspects', 'matched_aspects', 'similarity_score', 'predicted_sentiment']].head())
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from collections import Counter
from transformers import BertTokenizer, BertModel, BertForSequenceClassification
import torch
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
import pandas as pd
from gensim.models import Word2Vec
from IPython.display import display, HTML

#  1:    NLTK
nltk.download('stopwords')
nltk.download('wordnet')

#  2:   
stop_words = set(stopwords.words('russian'))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)  #  
    words = text.split()
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    return ' '.join(words)

#  3:    BERT
model_name = 'DeepPavlov/rubert-base-cased'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)
sentiment_model = BertForSequenceClassification.from_pretrained('blanchefort/rubert-base-cased-sentiment')

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
sentiment_model.to(device)

#  4:      BERT
def get_embedding(text):
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=128)
    inputs = {key: val.to(device) for key, val in inputs.items()}
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).cpu().numpy().flatten()

#  5:     Word2Vec
def train_word2vec(df):
    sentences = [review.split() for review in df['processed_review']]
    model = Word2Vec(sentences, vector_size=100, window=5, min_count=2, workers=4)
    return model

#  6:      Word2Vec
def phrase_similarity(phrase, word2vec_model, target_word):
    words = phrase.split()
    similarities = []
    for word in words:
        if word in word2vec_model.wv:
            similarities.append(word2vec_model.wv.similarity(word, target_word))
    if similarities:
        return np.mean(similarities)
    else:
        return 0

def filter_non_informative(aspects, word2vec_model, threshold=0.3):  #   
    informative_aspects = []
    for aspect in aspects:
        quality_sim = phrase_similarity(aspect, word2vec_model, '')
        works_sim = phrase_similarity(aspect, word2vec_model, '')
        if quality_sim > threshold or works_sim > threshold:
            informative_aspects.append(aspect)
    return informative_aspects

#  7:     TF-IDF    Word2Vec
def extract_aspects_tfidf(df, text_column='processed_review', num_terms=5):
    vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1, 3), stop_words=stop_words)  #   n-
    tfidf_matrix = vectorizer.fit_transform(df[text_column])
    feature_names = vectorizer.get_feature_names_out()

    aspects = {}
    
    for cluster in df['aspect_cluster'].unique():
        cluster_reviews = df[df['aspect_cluster'] == cluster][text_column]
        cluster_matrix = vectorizer.transform(cluster_reviews)
        mean_tfidf = cluster_matrix.mean(axis=0).A1
        top_indices = mean_tfidf.argsort()[-num_terms:][::-1]
        top_terms = [feature_names[i] for i in top_indices]
        aspects[cluster] = top_terms
    
    return aspects

#  8:       Word2Vec
def cluster_aspects(df, num_clusters=5):
    embeddings = np.array([get_embedding(review) for review in df['processed_review']])
    
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    df['aspect_cluster'] = kmeans.fit_predict(embeddings)

    word2vec_model = train_word2vec(df)
    aspects = extract_aspects_tfidf(df)
    
    #    Word2Vec
    filtered_aspects = {cluster: filter_non_informative(terms, word2vec_model) for cluster, terms in aspects.items()}
    
    #    DataFrame   
    aspect_clusters = pd.DataFrame(list(filtered_aspects.items()), columns=['aspect_cluster', 'common_terms'])
    
    return aspect_clusters, df

#  9:     
def predict_sentiment(text, rating):
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=128).to(device)
    with torch.no_grad():
        outputs = sentiment_model(**inputs)
        sentiment = torch.argmax(outputs.logits, dim=-1).cpu().numpy()
    
    #   
    if rating >= 4 and sentiment == 0:
        return 'positive'
    return 'positive' if sentiment == 1 else 'negative'

#  10:    
def analyze_sentiments(df):
    df['sentiment'] = df.apply(lambda row: predict_sentiment(row['processed_review'], row['review_rating']), axis=1)
    aspect_summary = df.groupby(['aspect_cluster', 'sentiment'])['processed_review'].count().unstack(fill_value=0)
    
    #    'positive'  'negative'
    if 'positive' not in aspect_summary.columns:
        aspect_summary['positive'] = 0
    if 'negative' not in aspect_summary.columns:
        aspect_summary['negative'] = 0
        
    aspect_summary['total'] = aspect_summary.sum(axis=1)
    return aspect_summary.sort_values('total', ascending=False)

#  11:    
df = df_raw[3000:3300]  #  df_raw   DataFrame  
df['processed_review'] = df['review_full_text'].apply(preprocess_text)
aspect_clusters, df = cluster_aspects(df, num_clusters=5)
aspect_summary = analyze_sentiments(df)

#  12:     
def prepare_presentation(aspect_clusters, aspect_summary):
    display(HTML("<h2>  </h2>"))
    
    for i, row in aspect_clusters.iterrows():
        aspect_terms = ', '.join(row['common_terms'])
        display(HTML(f"<h3> {row['aspect_cluster']}</h3>"))
        display(HTML(f"<p><strong> :</strong> {aspect_terms}</p>"))
        
        cluster_summary = aspect_summary.loc[[row['aspect_cluster']]]
        display(cluster_summary.style.set_caption("   ").background_gradient(cmap='coolwarm'))
    
    display(HTML("<h2></h2>"))
    rec_text = ("<p>     . "
                "      ,         .</p>"
                "<p>     ,    .</p>")
    display(HTML(rec_text))

#  13:   
prepare_presentation(aspect_clusters, aspect_summary)
```

```python
#   
import pandas as pd
import re
import time
from tqdm import tqdm  # -
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import spacy
from textblob import TextBlob
from transformers import pipeline, AutoTokenizer, AutoModel, AutoModelForSequenceClassification
from datasets import Dataset  #    Hugging Face Datasets
import gensim
from gensim.corpora import Dictionary
from gensim.models import LdaModel
from bertopic import BERTopic
import matplotlib.pyplot as plt
import seaborn as sns
import stanza

#   -     NLTK
nltk_stop_words = stopwords.words('russian')

#    CSV-,  cuDF
df_raw = pd.read_csv("./wildberries_reviews.csv.gz", nrows=30000, compression='gzip')
data = df_raw[3000:3500]

#      

# NLTK
def preprocess_text_nltk(text):
    text = re.sub(r'\W', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    tokens = word_tokenize(text.lower())
    stop_words = set(stopwords.words('russian'))
    tokens = [word for word in tokens if word not in stop_words]
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    return ' '.join(tokens)

tqdm.pandas()  #   
data['cleaned_text_nltk'] = data['review_full_text'].progress_apply(preprocess_text_nltk)

# SpaCy
nlp_spacy = spacy.load("ru_core_news_lg")  # large  spaCy
def preprocess_text_spacy(text):
    doc = nlp_spacy(text)
    tokens = [token.lemma_ for token in doc if not token.is_stop and not token.is_punct]
    return ' '.join(tokens)

data['cleaned_text_spacy'] = data['review_full_text'].progress_apply(preprocess_text_spacy)

# Stanza
nlp_stanza = stanza.Pipeline("ru", processors="tokenize,lemma")
def preprocess_text_stanza(text):
    doc = nlp_stanza(text)
    tokens = [word.lemma for sent in doc.sentences for word in sent.words if word.text not in stop_words]
    return ' '.join(tokens)

data['cleaned_text_stanza'] = data['review_full_text'].progress_apply(preprocess_text_stanza)

#   Hugging Face   
dataset = Dataset.from_pandas(data[['review_full_text', 'cleaned_text_spacy']])

# Hugging Face (  `rubert-base-cased`)
tokenizer_hf = AutoTokenizer.from_pretrained("DeepPavlov/rubert-base-cased")  # Base 
model_hf = AutoModelForSequenceClassification.from_pretrained("DeepPavlov/rubert-base-cased")
hf_pipeline = pipeline("sentiment-analysis", model=model_hf, tokenizer=tokenizer_hf, device=0)  #   GPU

#        Hugging Face   
candidate_labels = ["", "", "", ""]
aspect_pipeline = pipeline("zero-shot-classification", model="facebook/bart-large-mnli", device=0)  #   GPU

#     
def extract_aspects_batch(batch):
    texts = batch['cleaned_text_spacy']
    batch_aspects = []
    for text in texts:
        if not text.strip():
            batch_aspects.append([])
        else:
            result = aspect_pipeline(text, candidate_labels=candidate_labels)
            if not isinstance(result, list):
                result = [result]  #   
            batch_aspects.append(result)
    return {'aspects_hf': batch_aspects}

#   
dataset = dataset.map(extract_aspects_batch, batched=True, batch_size=8)


#     DataFrame
data['aspects_hf'] = dataset['aspects_hf']

#     TextBlob
data['sentiment_textblob'] = data['cleaned_text_nltk'].progress_apply(lambda x: TextBlob(x).sentiment.polarity)

#       rubert-base-cased
def sentiment_analysis_batch(batch):
    return {'sentiment_rubert': hf_pipeline(batch['review_full_text'])}

dataset = dataset.map(sentiment_analysis_batch, batched=True, batch_size=8)
data['sentiment_rubert'] = dataset['sentiment_rubert']

# Topic Modeling   Gensim LDA
#    -   
russian_stop_words = stopwords.words('russian')
vectorizer = CountVectorizer(max_df=0.95, min_df=2, stop_words=russian_stop_words)
dtm = vectorizer.fit_transform(data['cleaned_text_nltk'])
lda_model = LatentDirichletAllocation(n_components=5, random_state=42)
lda_model.fit(dtm)

#    
def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print(f"Topic {topic_idx}:")
        print(" ".join([feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]]))

display_topics(lda_model, vectorizer.get_feature_names_out(), 10)

#   `DeepPavlov/rubert-base-cased`     GPU
tokenizer = AutoTokenizer.from_pretrained("DeepPavlov/rubert-base-cased")
model = AutoModel.from_pretrained("DeepPavlov/rubert-base-cased").to("cuda")  #    GPU

#      GPU
def embed_texts(texts):
    inputs = tokenizer(texts, return_tensors="pt", padding=True, truncation=True).to("cuda")
    with torch.no_grad():
        embeddings = model(**inputs).last_hidden_state.mean(dim=1)
    return embeddings.cpu().numpy()  #    CPU    BERTopic

#    
embeddings = embed_texts(data['cleaned_text_spacy'].tolist())

#   BERTopic   -
vectorizer_model = CountVectorizer(stop_words=nltk_stop_words)
topic_model = BERTopic(language="russian", vectorizer_model=vectorizer_model)

#   BERTopic  
topics, probs = topic_model.fit_transform(data['cleaned_text_spacy'], embeddings)


#  -10 
print("Top 10 topics from BERTopic:")
print(topic_model.get_topic_info().head(10))

#   
def time_function(func, *args):
    start = time.time()
    func(*args)
    end = time.time()
    return end - start

# ,    0
if len(data) > 0:
    nltk_time = time_function(preprocess_text_nltk, data['review_full_text'].iloc[0])  #  iloc   
    spacy_time = time_function(preprocess_text_spacy, data['review_full_text'].iloc[0])
    stanza_time = time_function(preprocess_text_stanza, data['review_full_text'].iloc[0])

    display(f"NLTK processing time: {nltk_time:.4f} seconds")
    display(f"SpaCy processing time: {spacy_time:.4f} seconds")
    display(f"Stanza processing time: {stanza_time:.4f} seconds")
else:
    display("Dataframe is empty or has no rows with index 0")

#   
if len(data) > 0:
    display(data.head())
else:
    display("No data available for displaying results")
```

```python
import pandas as pd
import re
import spacy
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from tqdm import tqdm
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from transformers import pipeline, AutoTokenizer, AutoModelForSequenceClassification
from bertopic import BERTopic
import torch

#  
df_raw = pd.read_csv("./wildberries_reviews.csv.gz", nrows=30000, compression='gzip')
data = df_raw[3000:3500]

#  
def clean_text(text):
    text = re.sub(r'\n', ' ', text)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d+', '', text)
    text = text.lower()
    return text

data['cleaned_text'] = data['review_full_text'].apply(clean_text)

#    
grouped_data = data.groupby('product')['cleaned_text'].apply(lambda x: ' '.join(x)).reset_index()

#    -
stop_words = set(stopwords.words('russian'))
grouped_data['tokens'] = grouped_data['cleaned_text'].apply(lambda x: [word for word in word_tokenize(x) if word not in stop_words])

#    spaCy
nlp = spacy.load("ru_core_news_lg")

def lemmatize_text(tokens):
    doc = nlp(' '.join(tokens))
    return [token.lemma_ for token in doc]

tqdm.pandas()
grouped_data['lemmatized_tokens'] = grouped_data['tokens'].progress_apply(lemmatize_text)

#     Latent Dirichlet Allocation (LDA)
vectorizer = CountVectorizer(max_df=0.85, min_df=5, stop_words=stop_words)
lda_data = vectorizer.fit_transform(grouped_data['cleaned_text'])
lda_model = LatentDirichletAllocation(n_components=5, random_state=0)
lda_topics = lda_model.fit_transform(lda_data)

#    grouped_data
grouped_data['lda_topics'] = lda_topics.argmax(axis=1)

#     BERTopic
bertopic_model = BERTopic(language="russian")
topics, _ = bertopic_model.fit_transform(grouped_data['cleaned_text'])
grouped_data['bertopic'] = topics

# TF-IDF    
tfidf_vectorizer = TfidfVectorizer(max_features=20)
tfidf_matrix = tfidf_vectorizer.fit_transform(grouped_data['cleaned_text'])

#       
feature_names = tfidf_vectorizer.get_feature_names_out()

def get_top_keywords(tfidf_row, feature_names):
    top_indices = tfidf_row.argsort()[-10:][::-1]
    return [feature_names[i] for i in top_indices]

grouped_data['tfidf_keywords'] = [get_top_keywords(row, feature_names) for row in tfidf_matrix.toarray()]

# -    Hugging Face
device = 0 if torch.cuda.is_available() else -1
tokenizer_hf = AutoTokenizer.from_pretrained("blanchefort/rubert-base-cased-sentiment")
model_hf = AutoModelForSequenceClassification.from_pretrained("blanchefort/rubert-base-cased-sentiment")
hf_pipeline = pipeline("sentiment-analysis", model=model_hf, tokenizer=tokenizer_hf, device=device)

#      
def process_text_with_pipeline(text, pipeline):
    tokens = tokenizer_hf.encode(text, truncation=True, max_length=512)
    max_length = 128
    results = []
    
    for i in range(0, len(tokens), max_length):
        chunk = tokens[i:i + max_length]
        if len(chunk) < max_length:
            chunk += [tokenizer_hf.pad_token_id] * (max_length - len(chunk))
        text_chunk = tokenizer_hf.decode(chunk, skip_special_tokens=True)
        result = pipeline(text_chunk)[0]['label']
        results.append(result)
    
    sentiment_map = {"LABEL_1": "", "LABEL_0": "", "LABEL_2": ""}
    dominant_sentiment = max(set(results), key=results.count)
    return sentiment_map.get(dominant_sentiment, "")

grouped_data['hf_sentiment'] = grouped_data['cleaned_text'].apply(lambda x: process_text_with_pipeline(x, hf_pipeline))

#       LDA, BERTopic, TF-IDF  
def generate_key_thoughts(row):
    tfidf_keywords = row['tfidf_keywords']
    lda_topic = row['lda_topics']
    bertopic_topic = row['bertopic']
    hf_sentiment = row['hf_sentiment']
    
    key_thoughts = f"  :\n : {', '.join(map(str, tfidf_keywords))}\nLDA : {lda_topic}\nBERTopic : {bertopic_topic}\nSentiment: {hf_sentiment}"
    
    positive_aspects = []
    negative_aspects = []
    
    if hf_sentiment == "":
        positive_aspects.append("   ")
    if hf_sentiment == "":
        negative_aspects.append("   ")
    
    return key_thoughts, positive_aspects, negative_aspects

grouped_data['key_thoughts'], grouped_data['positive_aspects'], grouped_data['negative_aspects'] = zip(*grouped_data.apply(generate_key_thoughts, axis=1))

#  
for index, row in grouped_data.iterrows():
    print(f": {row['product']}")
    print(f" : {row['key_thoughts']}")
    print(" :")
    for aspect in row['positive_aspects']:
        print(f"- {aspect}")
    print(" :")
    for aspect in row['negative_aspects']:
        print(f"- {aspect}")
    print("\n")
```

```python
import spacy
from rake_nltk import Rake
from sklearn.cluster import KMeans
from sentence_transformers import SentenceTransformer
from transformers import AutoModel, AutoTokenizer
import torch
import numpy as np

#  
nlp = spacy.load("ru_core_news_lg")
model = SentenceTransformer('distiluse-base-multilingual-cased-v2')
tokenizer = AutoTokenizer.from_pretrained("DeepPavlov/rubert-base-cased")
transformer_model = AutoModel.from_pretrained("DeepPavlov/rubert-base-cased")

def extract_entities(text):
    doc = nlp(text)
    entities = [(ent.text, ent.label_) for ent in doc.ents]
    return entities

def generate_entity_based_summary(entities):
    entity_summary = " ".join([ent[0] for ent in entities])
    return f" : {entity_summary}"

def extract_key_phrases(text):
    r = Rake()
    r.extract_keywords_from_text(text)
    return r.get_ranked_phrases()

def generate_key_phrase_summary(phrases):
    return f" : {', '.join(phrases[:3])}"

def cluster_sentences(text):
    sentences = [s.strip() for s in text.split('.') if s.strip()]
    n_clusters = min(len(sentences), 3)
    
    if len(sentences) > 1:
        embeddings = model.encode(sentences)
        kmeans = KMeans(n_clusters=n_clusters)
        kmeans.fit(embeddings)
        cluster_sentences = [sentences[i] for i in kmeans.predict(embeddings)]
        return ". ".join(cluster_sentences)
    else:
        return text

def get_semantic_representation(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=512, return_overflowing_tokens=True)
    all_outputs = []
    for i in range(len(inputs['input_ids'])):
        input_slice = {k: v[i:i+1] for k, v in inputs.items() if k in ['input_ids', 'attention_mask', 'token_type_ids']}
        outputs = transformer_model(**input_slice)
        all_outputs.append(torch.mean(outputs.last_hidden_state, dim=1).detach().numpy())
    final_representation = np.mean(all_outputs, axis=0)
    return final_representation

def generate_semantic_summary(text):
    return " ".join(text.split()[:20])

#  
grouped_data['entities'] = grouped_data['cleaned_text'].apply(extract_entities)
grouped_data['entity_summary'] = grouped_data['entities'].apply(generate_entity_based_summary)
grouped_data['key_phrases'] = grouped_data['cleaned_text'].apply(extract_key_phrases)
grouped_data['key_phrase_summary'] = grouped_data['key_phrases'].apply(generate_key_phrase_summary)
grouped_data['clustered_summary'] = grouped_data['cleaned_text'].apply(cluster_sentences)
grouped_data['semantic_representation'] = grouped_data['cleaned_text'].apply(get_semantic_representation)
grouped_data['semantic_summary'] = grouped_data['cleaned_text'].apply(generate_semantic_summary)

#   
def generate_final_summary(row):
    return f"{row['entity_summary']}. {row['key_phrase_summary']}. {row['clustered_summary']}"

grouped_data['final_summary'] = grouped_data.apply(generate_final_summary, axis=1)

#  DataFrame    
comparison_df = grouped_data[['product', 'final_summary', 'entity_summary', 'key_phrase_summary', 'clustered_summary', 'semantic_summary']]
comparison_df.head(10)
```
