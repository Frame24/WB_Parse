```python
import pandas as pd
import json

# Функция для загрузки данных из JSON-файла и преобразования их в DataFrame
def load_json_to_df(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
        
    df = pd.DataFrame(columns=['Category', 'Subcategory', 'Characteristic', 'Synonyms', 'Units'])
    
    for category in data['categories']:
        category_name = category['category_name']
        
        for subcategory in category['subcategories']:
            subcategory_name = subcategory['subcategory_name']
            
            for characteristic in subcategory['characteristics']:
                characteristic_name = characteristic['name']
                synonyms = ', '.join(characteristic['synonyms'])  # Объединение синонимов в строку
                units = ', '.join(characteristic['units'])  # Объединение единиц измерения в строку
                
                # Добавление строки в DataFrame
                df = df.append({
                    'Category': category_name,
                    'Subcategory': subcategory_name,
                    'Characteristic': characteristic_name,
                    'Synonyms': synonyms,
                    'Units': units
                }, ignore_index=True)
    
    return df

# Загрузка данных из обоих файлов
df1 = load_json_to_df('./reviews_keywords/categories.json')
df2 = load_json_to_df('./reviews_keywords/categories2.json')

# Объединение двух DataFrame
combined_df = pd.concat([df1, df2], ignore_index=True)

# Вывод результата
print(combined_df.head())  # Показать первые 5 строк объединенного DataFrame

# Сохранение объединенного DataFrame в CSV файл, если необходимо
combined_df.to_csv('./reviews_keywords/combined_categories_dataframe.csv', index=False)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min
from IPython.display import display

# Инициализация TF-IDF векторизатора
vectorizer = TfidfVectorizer(max_features=1000)
X = vectorizer.fit_transform(df['processed_review'])  # Данные уже в pandas, не требуется преобразование

# Применение кластеризации для поиска новых аспектов
num_clusters = 10  # Количество кластеров можно настроить
kmeans = KMeans(n_clusters=num_clusters, random_state=0)
kmeans.fit(X)

# Получение новых аспектов из кластеров
new_aspects = []
for i in tqdm(range(num_clusters), desc="Извлечение новых аспектов"):
    cluster_center = kmeans.cluster_centers_[i]
    closest, _ = pairwise_distances_argmin_min(cluster_center.reshape(1, -1), X)
    aspect_candidate = df['processed_review'].iloc[closest[0]]
    
    # Убираем пустые строки и проверяем на наличие значимого текста
    if aspect_candidate and aspect_candidate.strip():
        new_aspects.append(aspect_candidate)

# Убираем дубликаты, если они есть
new_aspects = list(set(new_aspects))

display("Найденные новые аспекты:")
display(new_aspects)
```

```python
import json
import os
from datetime import datetime

# Пути для сохранения аспектов
new_aspects_file = './new_aspects.json'
blacklist_file = './aspect_blacklist.json'

# Функция для сохранения аспектов в файл
def save_aspects(aspects, file_path):
    with open(file_path, 'w', encoding='utf-8') as file:
        json.dump(aspects, file, ensure_ascii=False, indent=4)

# Функция для загрузки аспектов из файла
def load_aspects(file_path):
    if os.path.exists(file_path):
        with open(file_path, 'r', encoding='utf-8') as file:
            return json.load(file)
    return []

# Инициализация списка для новых аспектов
new_aspects_list = load_aspects(new_aspects_file)

# Проверка типа загруженных данных
if not isinstance(new_aspects_list, list):
    new_aspects_list = []

# Добавление новых аспектов в список (без дубликатов)
for aspect in new_aspects:
    if aspect not in new_aspects_list:
        new_aspects_list.append(aspect)

# Сохранение новых аспектов в файл
save_aspects(new_aspects_list, new_aspects_file)

# Функция для обзора аспектов и их сортировки в основной словарь или черный список
def review_and_sort_aspects(auto_confirm=False):
    new_aspects = load_aspects(new_aspects_file)
    blacklist = load_aspects(blacklist_file)
    
    updated_aspect_dict = aspect_dict.copy()  # Копируем основной словарь

    # Убедимся, что раздел 'Новые аспекты' существует
    if 'Новые аспекты' not in updated_aspect_dict:
        updated_aspect_dict['Новые аспекты'] = []

    for aspect in new_aspects:
        # Фильтрация аспектов по длине (например, минимум 3 слова)
        if len(aspect.split()) < 3:
            continue
        
        print(f"Новый аспект: {aspect}")
        
        if auto_confirm:
            action = 'a'  # Автоматическое добавление в словарь
        else:
            action = input("Добавить в словарь (a), добавить в черный список (b), пропустить (s): ").lower().strip()
        
        if action == 'a':
            updated_aspect_dict['Новые аспекты'].append(aspect)
        elif action == 'b':
            blacklist.append(aspect)
    
    # Сохраняем обновленный словарь аспектов и черный список
    save_aspects(updated_aspect_dict, new_aspects_file)
    save_aspects(blacklist, blacklist_file)
    print("Обзор завершен. Обновленные аспекты сохранены.")

# Вызов функции для обзора и сортировки с авто-подтверждением
review_and_sort_aspects(auto_confirm=True)
```

```python
import matplotlib.pyplot as plt
from transformers import BertTokenizer, BertForSequenceClassification
from transformers import TrainerCallback, TrainingArguments, Trainer
from torch.utils.data import Dataset
import torch
import pandas as pd
from IPython.display import display, update_display

# Загрузка предобученной модели и токенизатора BERT
model_name = 'DeepPavlov/rubert-base-cased'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForSequenceClassification.from_pretrained(model_name, num_labels=2)  # num_labels = 2 для двух классов

# Проверка доступности GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Перенос модели на GPU
model = model.to(device)
```

```python
# Кастомный класс Dataset
class ReviewsDataset(Dataset):
    def __init__(self, texts, labels):
        self.encodings = tokenizer(texts, truncation=True, padding=True, max_length=128)
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

# Пример данных для дообучения
texts = df['processed_review'].tolist()

# Преобразование меток
def convert_labels(rating):
    if rating < 4:
        return 0  # Негативный
    else:
        return 1  # Позитивный

labels = df['review_rating'].apply(convert_labels).tolist()  # Преобразуем метки в диапазон [0, 1]

# Создание датасета
dataset = ReviewsDataset(texts, labels)

# Настройка параметров обучения
training_args = TrainingArguments(
    output_dir='./reviews_keywords/results',
    num_train_epochs=5,
    per_device_train_batch_size=64,  # Устанавливаем размер батча на 64
    save_steps=10_000,
    save_total_limit=2,
    logging_steps=10,  # Логирование каждые 10 шагов
    logging_dir='./reviews_keywords/logs',
    fp16=True,  # Включение смешанного точного обучения
    report_to=[]  # Отключаем другие типы логирования
)

# Кастомный Trainer с переопределением метода сохранения
class CustomTrainer(Trainer):
    def save_model(self, output_dir=None, _internal_call=False):
        # Преобразование тензоров в непрерывные блоки памяти
        for param in self.model.parameters():
            param.data = param.data.contiguous()
        # Вызов стандартного метода сохранения модели
        super().save_model(output_dir, _internal_call)

# Инициализация кастомного Trainer с нашим callback
trainer = CustomTrainer(
    model=model,
    args=training_args,
    train_dataset=dataset,
)

# Запуск дообучения
train_output = trainer.train()
train_output
```

```python
import torch
from transformers import BertTokenizer, BertModel, BertForSequenceClassification
from sklearn.metrics.pairwise import cosine_similarity
from tqdm import tqdm
import numpy as np

# Загрузка предобученной модели и токенизатора
model_name = 'DeepPavlov/rubert-base-cased'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)
sentiment_model = BertForSequenceClassification.from_pretrained('./reviews_keywords/results/checkpoint-235')

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
sentiment_model.to(device)

# Функция для получения эмбеддинга текста
def get_embedding(text):
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=128)
    inputs = {key: val.to(device) for key, val in inputs.items()}
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).cpu().numpy().flatten()  # Используем .flatten() для преобразования вектора

# Функция для нормализации вектора
def normalize_vector(vector):
    norm = np.linalg.norm(vector)
    return vector / norm if norm != 0 else vector

# Функция для поиска ближайшей группы по косинусному расстоянию
def find_best_groups(category, aspect_groups, threshold=0.4):
    category_vector = normalize_vector(get_embedding(category))
    best_similarity = 0
    best_group = None
    
    for group_name in aspect_groups.keys():
        group_vector = normalize_vector(get_embedding(group_name))
        similarity = np.dot(category_vector, group_vector)  # Теперь вектора имеют правильную форму
        if similarity > best_similarity and similarity >= threshold:
            best_similarity = similarity
            best_group = group_name
    
    return ['основные характеристики', best_group] if best_group else ['основные характеристики']

# Функция для поиска ближайшего аспекта
def find_closest_aspect(embedding, aspect_embeddings):
    similarities = {aspect: cosine_similarity(embedding.reshape(1, -1), aspect_emb.reshape(1, -1)).flatten()[0] 
                    for aspect, aspect_emb in aspect_embeddings.items()}
    closest_aspect = max(similarities, key=similarities.get)
    return closest_aspect, similarities[closest_aspect]

# Функция для предсказания тональности
def predict_sentiment(text):
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=128).to(device)
    with torch.no_grad():
        outputs = sentiment_model(**inputs)
        sentiment = torch.argmax(outputs.logits, dim=-1).cpu().numpy()
    return 'positive' if sentiment == 1 else 'negative'

# Применение к DataFrame
def process_reviews(df, aspect_groups_all):
    categories = df['category'].unique()
    
    # Инициализируем кэш для аспектов
    aspect_embeddings_cache = {}
    
    # Результаты для каждого отзыва
    all_original_aspects = []
    all_matched_aspects = []
    all_similarity_scores = []
    all_sentiments = []

    for category in tqdm(categories, desc="Обработка категорий"):
        # Определение лучших групп для категории
        best_groups = find_best_groups(category, aspect_groups_all)
        
        # Создание эмбеддингов для аспектов из этих групп
        if category not in aspect_embeddings_cache:
            aspect_embeddings = {aspect: get_embedding(aspect) for group in best_groups for aspect in aspect_groups_all[group].keys()}
            aspect_embeddings_cache[category] = aspect_embeddings
        
        # Фильтруем датафрейм по категории
        category_df = df[df['category'] == category]
        
        for review in tqdm(category_df['processed_review'], desc=f"Обработка отзывов для категории {category}"):
            original_aspects = []
            matched_aspects = []
            similarity_scores = []
            sentiments = []

            # Разбиваем на фразы
            phrases = review.split('. ')
            
            for phrase in phrases:
                phrase_embedding = get_embedding(phrase)
                
                # Поиск ближайшего аспекта
                closest_aspect, similarity_score = find_closest_aspect(phrase_embedding, aspect_embeddings_cache[category])
                
                original_aspects.append(phrase)
                matched_aspects.append(closest_aspect)
                similarity_scores.append(similarity_score)
                sentiments.append(predict_sentiment(phrase))
            
            all_original_aspects.append(original_aspects)
            all_matched_aspects.append(matched_aspects)
            all_similarity_scores.append(similarity_scores)
            all_sentiments.append(sentiments)

    # Добавляем результаты в DataFrame
    df['original_aspects'] = all_original_aspects
    df['matched_aspects'] = all_matched_aspects
    df['similarity_score'] = all_similarity_scores
    df['predicted_sentiment'] = all_sentiments

    return df

# Применение процесса к части датафрейма
start_idx = 3000
end_idx = 6000
test_df = df_raw.iloc[start_idx:end_idx].copy()
test_df['processed_review'] = test_df['review_full_text'].apply(preprocess_text)

# Обработка и анализ отзывов
test_df = process_reviews(test_df, aspect_groups_all)

# Сохранение результатов
output_path = 'path_to_save_results.csv'
test_df.to_csv(output_path, index=False)
print(f"Результаты сохранены в {output_path}")

# Просмотр первых нескольких строк с результатами
display(test_df[['review_full_text', 'processed_review', 'original_aspects', 'matched_aspects', 'similarity_score', 'predicted_sentiment']].head())
```
