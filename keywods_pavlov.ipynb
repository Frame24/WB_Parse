```python
import pandas as pd
import json

#      JSON-     DataFrame
def load_json_to_df(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
        
    df = pd.DataFrame(columns=['Category', 'Subcategory', 'Characteristic', 'Synonyms', 'Units'])
    
    for category in data['categories']:
        category_name = category['category_name']
        
        for subcategory in category['subcategories']:
            subcategory_name = subcategory['subcategory_name']
            
            for characteristic in subcategory['characteristics']:
                characteristic_name = characteristic['name']
                synonyms = ', '.join(characteristic['synonyms'])  #    
                units = ', '.join(characteristic['units'])  #     
                
                #    DataFrame
                df = df.append({
                    'Category': category_name,
                    'Subcategory': subcategory_name,
                    'Characteristic': characteristic_name,
                    'Synonyms': synonyms,
                    'Units': units
                }, ignore_index=True)
    
    return df

#     
df1 = load_json_to_df('./reviews_keywords/categories.json')
df2 = load_json_to_df('./reviews_keywords/categories2.json')

#   DataFrame
combined_df = pd.concat([df1, df2], ignore_index=True)

#  
print(combined_df.head())  #   5   DataFrame

#   DataFrame  CSV ,  
combined_df.to_csv('./reviews_keywords/combined_categories_dataframe.csv', index=False)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min
from IPython.display import display

#  TF-IDF 
vectorizer = TfidfVectorizer(max_features=1000)
X = vectorizer.fit_transform(df['processed_review'])  #    pandas,   

#      
num_clusters = 10  #    
kmeans = KMeans(n_clusters=num_clusters, random_state=0)
kmeans.fit(X)

#     
new_aspects = []
for i in tqdm(range(num_clusters), desc="  "):
    cluster_center = kmeans.cluster_centers_[i]
    closest, _ = pairwise_distances_argmin_min(cluster_center.reshape(1, -1), X)
    aspect_candidate = df['processed_review'].iloc[closest[0]]
    
    #         
    if aspect_candidate and aspect_candidate.strip():
        new_aspects.append(aspect_candidate)

#  ,   
new_aspects = list(set(new_aspects))

display("  :")
display(new_aspects)
```

```python
import matplotlib.pyplot as plt
from transformers import BertTokenizer, BertForSequenceClassification
from transformers import TrainerCallback, TrainingArguments, Trainer
from torch.utils.data import Dataset
import torch
import pandas as pd
from IPython.display import display, update_display

#      BERT
model_name = 'DeepPavlov/rubert-base-cased'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForSequenceClassification.from_pretrained(model_name, num_labels=2)  # num_labels = 2   

#   GPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

#    GPU
model = model.to(device)
```

```python
#   Dataset
class ReviewsDataset(Dataset):
    def __init__(self, texts, labels):
        self.encodings = tokenizer(texts, truncation=True, padding=True, max_length=128)
        self.labels = labels

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

#    
texts = df['processed_review'].tolist()

#  
def convert_labels(rating):
    if rating < 4:
        return 0  # 
    else:
        return 1  # 

labels = df['review_rating'].apply(convert_labels).tolist()  #     [0, 1]

#  
dataset = ReviewsDataset(texts, labels)

#   
training_args = TrainingArguments(
    output_dir='./reviews_keywords/results',
    num_train_epochs=5,
    per_device_train_batch_size=64,  #     64
    save_steps=10_000,
    save_total_limit=2,
    logging_steps=10,  #   10 
    logging_dir='./reviews_keywords/logs',
    fp16=True,  #    
    report_to=[]  #    
)

#  Trainer    
class CustomTrainer(Trainer):
    def save_model(self, output_dir=None, _internal_call=False):
        #      
        for param in self.model.parameters():
            param.data = param.data.contiguous()
        #     
        super().save_model(output_dir, _internal_call)

#   Trainer   callback
trainer = CustomTrainer(
    model=model,
    args=training_args,
    train_dataset=dataset,
)

#  
train_output = trainer.train()
train_output
```

```python
import torch
from transformers import BertTokenizer, BertModel, BertForSequenceClassification
from sklearn.metrics.pairwise import cosine_similarity
from tqdm import tqdm
import numpy as np

#     
model_name = 'DeepPavlov/rubert-base-cased'
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name)
sentiment_model = BertForSequenceClassification.from_pretrained('./reviews_keywords/results/checkpoint-235')

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
sentiment_model.to(device)

#     
def get_embedding(text):
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=128)
    inputs = {key: val.to(device) for key, val in inputs.items()}
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).cpu().numpy().flatten()  #  .flatten()   

#    
def normalize_vector(vector):
    norm = np.linalg.norm(vector)
    return vector / norm if norm != 0 else vector

#        
def find_best_groups(category, aspect_groups, threshold=0.4):
    category_vector = normalize_vector(get_embedding(category))
    best_similarity = 0
    best_group = None
    
    for group_name in aspect_groups.keys():
        group_vector = normalize_vector(get_embedding(group_name))
        similarity = np.dot(category_vector, group_vector)  #     
        if similarity > best_similarity and similarity >= threshold:
            best_similarity = similarity
            best_group = group_name
    
    return [' ', best_group] if best_group else [' ']

#     
def find_closest_aspect(embedding, aspect_embeddings):
    similarities = {aspect: cosine_similarity(embedding.reshape(1, -1), aspect_emb.reshape(1, -1)).flatten()[0] 
                    for aspect, aspect_emb in aspect_embeddings.items()}
    closest_aspect = max(similarities, key=similarities.get)
    return closest_aspect, similarities[closest_aspect]

#    
def predict_sentiment(text):
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=128).to(device)
    with torch.no_grad():
        outputs = sentiment_model(**inputs)
        sentiment = torch.argmax(outputs.logits, dim=-1).cpu().numpy()
    return 'positive' if sentiment == 1 else 'negative'

#   DataFrame
def process_reviews(df, aspect_groups_all):
    categories = df['category'].unique()
    
    #    
    aspect_embeddings_cache = {}
    
    #    
    all_original_aspects = []
    all_matched_aspects = []
    all_similarity_scores = []
    all_sentiments = []

    for category in tqdm(categories, desc=" "):
        #     
        best_groups = find_best_groups(category, aspect_groups_all)
        
        #       
        if category not in aspect_embeddings_cache:
            aspect_embeddings = {aspect: get_embedding(aspect) for group in best_groups for aspect in aspect_groups_all[group].keys()}
            aspect_embeddings_cache[category] = aspect_embeddings
        
        #    
        category_df = df[df['category'] == category]
        
        for review in tqdm(category_df['processed_review'], desc=f"    {category}"):
            original_aspects = []
            matched_aspects = []
            similarity_scores = []
            sentiments = []

            #   
            phrases = review.split('. ')
            
            for phrase in phrases:
                phrase_embedding = get_embedding(phrase)
                
                #   
                closest_aspect, similarity_score = find_closest_aspect(phrase_embedding, aspect_embeddings_cache[category])
                
                original_aspects.append(phrase)
                matched_aspects.append(closest_aspect)
                similarity_scores.append(similarity_score)
                sentiments.append(predict_sentiment(phrase))
            
            all_original_aspects.append(original_aspects)
            all_matched_aspects.append(matched_aspects)
            all_similarity_scores.append(similarity_scores)
            all_sentiments.append(sentiments)

    #    DataFrame
    df['original_aspects'] = all_original_aspects
    df['matched_aspects'] = all_matched_aspects
    df['similarity_score'] = all_similarity_scores
    df['predicted_sentiment'] = all_sentiments

    return df

#     
start_idx = 3000
end_idx = 6000
test_df = df_raw.iloc[start_idx:end_idx].copy()
test_df['processed_review'] = test_df['review_full_text'].apply(preprocess_text)

#    
test_df = process_reviews(test_df, aspect_groups_all)

#  
output_path = 'path_to_save_results.csv'
test_df.to_csv(output_path, index=False)
print(f"   {output_path}")

#      
display(test_df[['review_full_text', 'processed_review', 'original_aspects', 'matched_aspects', 'similarity_score', 'predicted_sentiment']].head())
```
